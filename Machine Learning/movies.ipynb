```python
#Importamos las librerias
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
#Se importa la data

peliculas = pd.read_csv("data/movies2.csv")
```

```python
peliculas
```

```python
#Seleccionar solo los valores numÃ©ricos
datos_numericos = peliculas.select_dtypes(np.number)

```

```python
datos_numericos
```

```python
# Reemplazamos los Nan por 0

datos_limpios = peliculas.select_dtypes(np.number).fillna(0)
```

```python
datos_limpios
```

```python
#Configurar las variables objetivo e independientes.

objetivo = "ventas"
#las variables independientes serian todas las demas menos ventas
independientes = datos_limpios.drop(columns=objetivo).columns
```

```python
#Creamos el modelo y lo ajustamos

modelo = LinearRegression()
modelo.fit(X=datos_limpios[independientes], y=datos_limpios[objetivo])
```

```python
#Agregamos las predicciones como una nueva columna del dataset original
peliculas["ventas_prediccion"] = modelo.predict(datos_limpios[independientes])
```
