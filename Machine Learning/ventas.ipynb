```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
ventas = pd.read_csv("data/ventas2.csv")
objetivo = "monto"
independientes = ventas.drop(columns=['monto']).columns
```

```python
ventas
```

```python
objetivo
```

```python
independientes
```

# Se crea el modelo

```python
modelo = LinearRegression()
modelo.fit(X=ventas[independientes], y=ventas[objetivo])
```

```python
#Creamos un conjunto solo con los datos ventas reales y ventas de predicción.
ventas["ventas_prediccion"] = modelo.predict(ventas[independientes])
preds = ventas[["monto", "ventas_prediccion"]].head(50)
```

```python
#Realizamos una predicción 
talvez = modelo.predict([[41,1,1,1]])
print ("Tal vez compre: $")
talvez
#Imagine que viene una persona de 41 años, solo, en su propio vehículo y pagando en efectivo. 
#Entonces la lista sería: [41,1,1,1]Observamos que una persona con esas características compraría: $1,139.31
```
